# 谐和激励

$$
\begin{equation}
    \ddot{x} + 2 \xi \omega_0 \dot{x} + \omega_0^2 x = \frac{F_0}{m} \cos(\omega_0 t + \varphi_0)
\end{equation}
$$

得到振动解

$$
\begin{equation}
    \tilde{x}(t) = \beta(\eta) \frac{F_0}{k} e^{\mathrm{i} (\omega t - \varphi_0(\eta))}
\end{equation}
$$

其中

$$
\begin{aligned}
    \beta(\eta) &= \frac{1}{\sqrt{(1 - \eta^2)^2 + (2 \xi \eta)^2}} \\
    \varphi_0(\eta) &= \arctan \frac{2 \xi \eta}{1 - \eta^2}
\end{aligned}
$$

```python
# Plot beta(eta) and phi(eta)

import numpy as np
import matplotlib.pyplot as plt

def beta(eta, xi=0.5):
    return 1 / (np.sqrt((1 - eta**2)**2 + 4 * xi**2 * eta**2))

def phi(eta, xi=0.5):
    return np.arctan2(2 * xi * eta, 1 - eta**2)

eta = np.linspace(0, 3, 500)
xi = 0.01
beta_values = beta(eta, xi)
phi_values = phi(eta, xi)

plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(eta, beta_values, label=f'ξ={xi}')
plt.title('Frequency Response')
plt.ylabel('β(η)')
plt.grid()
plt.legend()
plt.subplot(2, 1, 2)
plt.plot(eta, phi_values, label=f'ξ={xi}')
plt.title('Phase Response')
plt.ylabel('φ(η)')
plt.grid()
plt.legend()
plt.tight_layout()
```
