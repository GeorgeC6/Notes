一维无限深方势阱的解：

$$
\begin{equation}
    \psi_n(x) = \sqrt{\frac{2}{a}} \sin\left(\frac{n \pi}{a} x\right), \, n = 1, 2, \ldots
\end{equation}
$$

```python
# 画出前三个能级的波函数，并排列展示
%config InlineBackend.figure_format = 'retina'

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import gridspec
from scipy.constants import hbar, pi
from scipy.integrate import simps
from scipy.special import erf
from scipy.optimize import fsolve
from IPython.display import display, Math
from matplotlib.patches import Rectangle
from matplotlib.ticker import MultipleLocator, AutoMinorLocator
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 12
rcParams['mathtext.fontset'] = 'cm'
rcParams['figure.dpi'] = 150
rcParams['axes.linewidth'] = 1.0
rcParams['lines.linewidth'] = 1.5
rcParams['lines.markersize'] = 6
rcParams['xtick.direction'] = 'in'
rcParams['ytick.direction'] = 'in'
rcParams['xtick.top'] = True
rcParams['ytick.right'] = True
rcParams['xtick.major.size'] = 5
rcParams['ytick.major.size'] = 5
rcParams['xtick.minor.size'] = 2.5
rcParams['ytick.minor.size'] = 2.5
rcParams['xtick.major.width'] = 1.0
rcParams['ytick.major.width'] = 1.0
rcParams['xtick.minor.width'] = 1.0
rcParams['ytick.minor.width'] = 1.0
rcParams['legend.frameon'] = False
rcParams['legend.fontsize'] = 10
rcParams['figure.figsize'] = (12, 4)

# 画图
fig = plt.figure(constrained_layout=True)
gs = gridspec.GridSpec(ncols=3, nrows=1, figure=fig)
ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1])
ax3 = fig.add_subplot(gs[0, 2])
axes = [ax1, ax2, ax3]
x = np.linspace(0, 1, 1000)
for n, ax in enumerate(axes):
    psi_n = np.sqrt(2) * np.sin((n + 1) * pi * x)
    ax.plot(x, psi_n, label=r"$\psi_{{{}}}(x)$".format(n + 1))
    ax.axhline(0, color='black', lw=0.5)
    ax.set_xlim(0, 1)
    ax.set_ylim(-2, 2)
    ax.set_yticks([-1, 0, 1])
    ax.set_xticks([0, 0.5, 1])
    ax.xaxis.set_minor_locator(AutoMinorLocator())
    ax.yaxis.set_minor_locator(AutoMinorLocator())
    ax.set_xlabel(r"$x$")
    ax.set_ylabel(r"$\psi_{{{}}}(x)$".format(n + 1))
    ax.legend()
```
